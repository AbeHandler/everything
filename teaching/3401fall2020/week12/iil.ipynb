```python
import pandas as pd 

# https://www.kaggle.com/pieca111/music-artists-popularity

df = pd.read_csv("artists.csv")

df = df.fillna(value="")

df = df[df["country_mb"] == "United States"]

print(len(df))

all_artists = df["artist_mb"].to_list()

all_artists = [o.lower() for o in all_artists]

artist2toks = {}

for d in tqdm(all_artists): 
    artist2toks[d] = set([k.lower() for k in d.split(" ") if k.lower() not in stops])
```
