```python
import pandas as pd

a = pd.DataFrame({'name': ["sir john", "fluffy", "mittens", "taco", "gwyneth"], 
                 "species": ["dog", "cat", "cat", "cat", "dog"]})
```

```python
# merge = columns
# join = indexes 
```

```python
b = pd.DataFrame({"date": ["2020-06-29 08:22:14", "2020-06-29 08:07:17",
                           "2020-06-29 08:23:33", "2020-06-28 08:02:35",
                           "2020-06-29 08:15:12", "2020-06-28 08:14:27"],
                  "name": ["fluffy", "mittens", "gwyneth", "fluffy", "mittens", "taco"]
                 })
```

```python
a.merge(b, how="inner")
```

```python
a.merge(b, how="outer", on="name")
```

```python
a.merge(b, how="left", on="name")
```

```python
a.merge(b, how="right", on="name")
```

```python
a.join(b, how="inner", lsuffix='_l', rsuffix='_r')
```

```python
a.join(b, how="outer", lsuffix='_l', rsuffix='_r')
```
