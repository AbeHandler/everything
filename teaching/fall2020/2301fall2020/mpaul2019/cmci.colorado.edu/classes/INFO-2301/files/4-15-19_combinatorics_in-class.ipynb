```python
breads = {'wheat', 'white'}
meats = {'none', 'ham', 'turkey'}
cheeses = {'cheddar', 'swiss', 'provolone', 'american'}
toppings = {'lettuce', 'tomatoes', 'onions', 'pickles'}
sauces = {'mustard', 'mayo'}
```

```python
# How many different ways can you make a sandwich with 1 ingredient from each category?
# 2 * 3 * 4 * 4 * 2 = 192

count = 0
for bread in breads:
    for meat in meats:
        for cheese in cheeses:
            for topping in toppings:
                for sauce in sauces:
                    count += 1
                    print(bread, meat, cheese, topping, sauce)
print("Number of permutations:", count)
```

```python
# How many different ways can you choose 2 cheeses out of the 4 options?
# 4 choose 2 = 6

count = 0
for cheese1 in cheeses:
    for cheese2 in cheeses:
        if cheese1 != cheese2 and cheese1 < cheese2:
            count += 1
            print(cheese1, cheese2)
print("Number of combinations:", count)
```
