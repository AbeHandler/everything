```python
from math import sqrt

def magnitude(x):
    total = 0
    for i in range(len(x)):
        total += pow(x[i], 2)   # x_i^2
    return sqrt(total)

def dot_product(x, y):
    total = 0
    for i in range(len(x)):
        total += (x[i]*y[i])
    return (total)

def cos_sim(x, y):
    return dot_product(x, y) / (magnitude(x) * magnitude(y))
```
