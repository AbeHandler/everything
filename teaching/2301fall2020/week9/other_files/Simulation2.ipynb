```python
import random

probOfAge = {"18-34": 2.4/5, "34+": 2.6/5}

pBidenGivenAge = {"18-34": 1.7/2.4,  "34+": 1.1/2.6}
```

```python
def drawAge(probOfAge):
    draw = random.random()
    if draw < probOfAge["18-34"]:
        return "18-34"
    else:
        return "34+"
    
def drawBiden(pBidenGivenAge, age):
    draw = random.random()
    if draw < pBidenGivenAge["18-34"]:
        return "Biden"
    else:
        return "Trump"
```

```python
from collections import Counter

def election(Nvoters):
    output = []
    for n in range(Nvoters):
        age = drawAge(probOfAge)
        vote = drawBiden(pBidenGivenAge, age)
        output.append(vote)
    return output

for simulated_election in range(1000):
    votes = election(Nvoters = 100)
    votecounts = Counter(votes)
    biden_won = votecounts["Biden"] > votecounts["Trump"]
```
